```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
wijkmigratie = pd.read_csv("Wijkmigratie.csv", sep=";", index_col=0)
wijkmigratie.columns = [ year[-4:] for year in wijkmigratie.columns]
```

```python
wijkmigratie.head()
```

```python
wijkmigratie['TOTAAL'] = wijkmigratie.sum(axis=1)
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
sns.heatmap(wijkmigratie, annot=True, fmt="d", linewidths=.5, 
            cmap="PiYG", ax=ax, vmin=-200, vmax=200)
plt.savefig('wijkmigratie_gent.pdf')
```
